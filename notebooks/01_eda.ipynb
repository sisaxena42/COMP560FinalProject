```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.model import load_raw_data, add_targets

df = load_raw_data("../data/merged.csv")
df = add_targets(df)
df.head()
```

```python
df.info()
df.describe(include="all").transpose().head(20)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df["GDP"], ax=axes[0])
axes[0].set_title("GDP distribution")

sns.histplot(df["GDP_per_capita"], ax=axes[1])
axes[1].set_title("GDP per capita distribution")

plt.tight_layout()
```

```python
numeric_cols = df.select_dtypes(include="number").columns
plt.figure(figsize=(10, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation matrix (numeric features)")
plt.show()
```
